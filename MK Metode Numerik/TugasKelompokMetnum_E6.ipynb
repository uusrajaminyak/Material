# GALAT
## Definisi Galat

Galat adalah kesalahan atau kekeliruan yang terjadi dalam pengambilan data, perhitungan, atau pengukuran. Galat juga bisa diartikan sebagai selisih antara nilai yang diharapkan dengan hasil yang sebenarnya.

## Deret MacLaurin (Taylor Baku)

Deret Maclaurin atau deret taylor baku  adalah kasus khusus dari deret Taylor dengan fungsi expansi di sekitar (X0) = 0.

Salah satu contoh kasus adalah menghitung nilai sin suatu sudut x. Pertama masukkan sin x ke dalam rumus deret Maclaurin seperti berikut.

Lalu persamaan tersebut dimasukkan dalam kode program untuk menghitung nilai sin secara analitik (hampiran)

```python
import math

def sin_taylor(x, n):
    # Menghitung sin(x) menggunakan deret Taylor hingga n suku.
    # x dalam radian.

    sin_hampiran = 0
    for n in range(n):
        rumus = ((-1)**n) * (x**(2*n + 1)) / math.factorial(2*n + 1)
        sin_hampiran += rumus
    return sin_hampiran
```

Program di bawah ini akan menghitung nilai sin secara numerik (sejati) dan nilai galat mutlak dan relatifnya.

```python
def galat_relatif(x, n):
    # Menghitung galat relatif antara pendekatan deret Taylor dan nilai eksak.

    sin_sejati = math.sin(x)
    sin_hampiran = sin_taylor(x, n)
    galat_mutlak = abs(sin_sejati - sin_hampiran)
    galat_relatif = (galat_mutlak / abs(sin_sejati)) * 100 if sin_sejati != 0 else None
    return sin_hampiran, galat_mutlak, galat_relatif
```

Jalankan program. Sebagai contoh perhitungan yaitu menghitung sin 30 dengan 4 suku deret.

```python
# Contoh perhitungan untuk x = /6 (30 derajat) dengan 4 suku
x = math.pi / 6
n = 4

sin_hampiran, galat_mutlak, galat_relatif = galat_relatif(x, n)

print(f"sin({x}) sejati       : {math.sin(x):.10f}")
print(f"sin({x}) hampiran     : {sin_hampiran:.10f}")
print(f"Galat absolut         : {galat_mutlak:.10f}")
print(f"Galat relatif (%)     : {galat_relatif:.10f}" if galat_relatif is not None else "Galat relatif tidakterdefinisi")
```
